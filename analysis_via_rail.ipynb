```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
via_data=pd.read_excel('via trail.xlsx')
via_data
```

```python
# List of columns to convert to numeric
numeric_columns = ['Revenues\n(in thousands)', 'Costs\n(in thousands)', 'Shortfall\n(in thousands)',
                   'Subsidy\nper passenger', 'Passengers\nper week', 'Passengers\nper year']

# Remove commas and convert selected columns to numeric
for col in numeric_columns:
    via_data[col] = via_data[col].str.replace(',', '').apply(pd.to_numeric, errors='coerce')
```

```python
via_data.dtypes
```

```python
# Replace NaN values with zeros in the entire DataFrame
via_data.fillna(0, inplace=True)
via_data
```

```python
# Calculate profitability for each train service
via_data['Profitability'] = (via_data['Revenues\n(in thousands)'] - via_data['Costs\n(in thousands)']) 

```

```python
# Calculate subsidy per destination for each train service
via_data['Total subsidy per destination'] = (via_data['Subsidy\nper passenger'] * via_data['Passengers\nper year'])
# Format and print the subsidy per destination amounts without scientific notation
formatted_subsidy = via_data['Total subsidy per destination'].apply(lambda x: "{:.2f}".format(x))
formatted_subsidy
```

```python
# Calculate subsidy percentage of costs and revenues for each train service
via_data['Subsidy % of Costs'] = (via_data['Total subsidy per destination'] / (via_data['Costs\n(in thousands)'] * 1000)) * 100
via_data['Subsidy % of Revenues'] = (via_data['Total subsidy per destination']  / (via_data['Revenues\n(in thousands)'] * 1000)) * 100
via_data
```

```python
via_data.dtypes
```

```python
# Sort the DataFrame by 'Subsidy % of Costs' in ascending order
via_data_sorted = via_data.sort_values(by='Subsidy % of Costs')

# Specify the columns to plot
x = 'Subsidy % of Costs'
y = 'Train Services'

# Create the matplotlib figure
plt.figure(figsize=(10, 6))

# Plot the data as a horizontal bar plot
plt.barh(via_data_sorted[y], via_data_sorted[x])

# Customize the plot
plt.title('Subsidy Percentages of Costs')
plt.xlabel('Percentage (%)')
plt.ylabel('Train Services')

# Show the plot
plt.tight_layout()



```

```python
# Create the hvplot figure
# Round to one decimal point and add '%' symbol
via_data['Subsidy % of Costs'] = via_data['Subsidy % of Costs'].apply(lambda x: f"{x:.1f}%")
via_data['Subsidy % of Revenues'] = via_data['Subsidy % of Revenues'].apply(lambda x: f"{x:.1f}%")

```
